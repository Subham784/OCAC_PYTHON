### scientific calculater

```python
import tkinter as tk
from math import *

# Function to update the input field
def click_button(event):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, current + event)

# Function to clear the input field
def clear_button():
    entry.delete(0, tk.END)

# Function to evaluate the input
def evaluate_button():
    try:
        result = eval(entry.get(), {"__builtins__": None}, globals())
        entry.delete(0, tk.END)
        entry.insert(tk.END, str(result))
    except Exception:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Function to create scientific buttons
def create_sci_buttons():
    sci_operations = [
        ('sin', 'sin('), ('cos', 'cos('), ('tan', 'tan('), ('log', 'log('),('sec','sec('),('cot','cot('),('cosec','cosec('),
        ('√', 'sqrt('), ('π', 'pi'), ('e', 'e'), ('^', '**')
    ]
    for i, (text, symbol) in enumerate(sci_operations):
        button = tk.Button(root, text=text, command=lambda symbol=symbol: click_button(symbol))
        button.grid(row=5 + i // 4, column=i % 4, sticky="nsew")

# Create the main window
root = tk.Tk()
root.title("Scientific Calculator")

# Create the entry widget
entry = tk.Entry(root, width=30, font=('Arial', 14), borderwidth=5)
entry.grid(row=0, column=0, columnspan=4)

# Create digit buttons
digits = [
    '7', '8', '9',
    '4', '5', '6',
    '1', '2', '3',
    '0', '.'
]
for i, digit in enumerate(digits):
    button = tk.Button(root, text=digit, command=lambda digit=digit: click_button(digit))
    button.grid(row=1 + i // 3, column=i % 3, sticky="nsew")

# Create basic operation buttons
operations = [
    ('+', '+'), ('-', '-'), ('*', '*'), ('/', '/')
]
for i, (text, symbol) in enumerate(operations):
    button = tk.Button(root, text=text, command=lambda symbol=symbol: click_button(symbol))
    button.grid(row=1 + i, column=3, sticky="nsew")

# Create special buttons
tk.Button(root, text='C', command=clear_button).grid(row=4, column=1, sticky="nsew")
tk.Button(root, text='=', command=evaluate_button).grid(row=4, column=2, sticky="nsew")
# tk.Button(root, text='~', command=backspace_button).grid(row=4,column=3, sticky="nsew")
# Create scientific function buttons
create_sci_buttons()

# Adjust the layout
for i in range(8):
    root.grid_rowconfigure(i, weight=1)
for i in range(4):
    root.grid_columnconfigure(i, weight=1)

# Run the application
root.mainloop()
```
