## MatplotLib

###### Matplotib is a plotting library for the Pyton programming language and its numerical mathematics extension Numpy.

###### Some of the major Pros of Matplotlib are:

###### .Generally easy to get started for simple plots
###### .Support for custom labels and texts
###### .Great control of very element in a fgure
###### . High-quality output in many formats
###### .Very customizable in general

###### .plt.bar():Make a bar plot.
###### .plt.plot():Can be used to make a line graph.
###### .plt.subplots():Add a subplot to the current figure.it returns two arguments fig and ax where fig is figured and ax can be either a single Axes objects or an Axes objects.
###### .plt.xticks():A list of position at which ticks should be placed.
###### .plt.xlabel(label_name_):Set the label for the x-asis.
###### .plt.ylabel(label_name_):Set the label for the y-asis.
###### .plt.show():Display a figure.
###### .plt.grid():Configure the grid lines.

## Refer to the https://matplotlib.org/stable/index.html for more details.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({'Length':np.random.randn(500)})
df.head()
```

### Histogram
###### in histogram x axis contains a variable and y axis will be a frequency of that variable

```python
df['Length'].hist()  #using pandas
plt.show()
```

```python
plt.hist(df['Length'])
plt.show()
```

```python
plt.hist(df['Length'],bins=15)
plt.show()
```

```python
plt.hist(df['Length'],bins=15)
plt.xlabel('Length.')
plt.ylabel('Count')
plt.title('Histogram of length')
plt.show()
```

```python
# horizontal orientation
plt.hist(df['Length'],bins=15, orientation = 'horizontal')
plt.xlabel('Length  in mm.')
plt.ylabel('Count')
plt.title('Histogram of length')
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,36]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

# set axes labels and chart title 
plt.xlabel('Day')
plt.ylabel('Tempreture')
plt.title('Weather')

# set legend
plt.legend()
plt.show()
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'average')

plt.legend(loc='upper right',shadow=True,fontsize='small')
plt.show()
```

## pie Chart

```python
list = [20,80,15,10,20,5]
lbs = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampo', 'Moisture']
```

```python
# Creating and displaying a pie chart.
plt.pie(list,labels=lbs)
plt.show()
```

```python
# Creating and displaying a pie chart.
plt.pie(list,labels=lbs,autopct='%1.1f%%')
plt.show()
```

```python
# Draw pie chart as perfect circle
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
# Draw pie chart as perfect circle 
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

```python
# Shaw percentage for every pie. Specify radius to increase chart
plt.pie(list,labels=lbs,autopct='%1.1f%%',shadow=True ,radius=1.5)
plt.axis('equal')
plt.show()
```

```python
# CVounterclock and angle properties
plt.pie(list,labels=lbs,shadow=True,autopct='%1.1f%%',radius=1.5,explode=[0,0,0,1.0,0,0])
```

```python
# Data to plot
la='python','C++','Ruby','java'
sizes=[215,130,205,210]
co = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)  # explode 1st slice

#plot
plt.pie(sizes, explode = explode, labels = la, colors = co,autopct='%1.1f%%')

plt.axis('equal')
plt.show()
```

## Bar plot

```python
# 5 students' marks are plotted on a bar graph.
marks_scored = [81,90,45,72,99]
roll_no = [11,12,13,14,15]
plt.bar(roll_no,marks_scored)
plt.xlabel('roll no')
plt.ylabel('marks scored')
plt.show()
```

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2, y2, color ='g')

plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
# Simple bar chsrt showing revenues of major us tech companies
company = ['GOOGLE','AMAZON','MSFT','FB']
revenue = [90,136,89,27]

plt.bar(company,revenue,label='Revenue')
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
# Horizontal bar chart using barh function
plt.barh(company,revenue,label='Revenure')
plt.ylabel('Revenue(Bln)')
plt.title('US Technology Stocks')
plt.legend()
```

```python
# Multiple Bars dshowing revenue and profit of major UsS tech companies
company=['GOOGLE','AMZON','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')

plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
# Multiple Bars showing revenue and profit of major US tech companies
company=['GOOGLE','AMZON','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos = np.arange(len(company))
plt.bar(xpos-0.2,revenue, width=0.4, label = 'Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.xticks(xpos,company)
plt.ylabel('Revenue(Bin)')
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
# Program for drawing multiplebar charts on one image.
ecommerce=['Myntra','Snapdeal','Alibaba','Amazon','Flipkart']
Q1_profit=[35, 45, 100, 70, 40]
Q2_profit=[38, 40, 105, 65, 45]
Q3_profit=[30, 42, 120, 72, 50]
Q4_profit=[25, 34, 115, 60, 48]

# Creating different bar charts on one image using subplot() function.
plt.figure(1,figsize=(10,10))

# Creating bar chart in first cell of figure having 2 rows, 3 columns
plt.subplot(221)
plt.bar(ecommerce,Q1_profit)
plt.title('Quarter1 Profit')

# Creating a bar chart in second cell
plt.subplot(222)
plt.bar(ecommerce,Q2_profit)
plt.title('Quarter2 Profit')

# Creating a bar chart in third cell
plt.subplot(223)
plt.bar(ecommerce,Q3_profit)
plt.title('Quarter3 Profit')

# Creating a bar chart in fourth cell
plt.subplot(224)
plt.bar(ecommerce,Q4_profit)
plt.title('Quarter4 Profit')

#Adding a maibn title for the figure
plt.suptitle('profit on quater basis')

# Display the chart
plt.show()
```
