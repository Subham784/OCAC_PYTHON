###### Concatenate

```python
india_weather = pd.DataFrame({
    'city':['mumbai','delhi','banglore'],
    'tempreture':[32,45,30],
    'humidity':[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'tempreture':[21,14,35],
    'humidity':[68,65,75]
    })
us_weather
```

```python
# concate two data frames
df = pd.concat([india_weather, us_weather])
df
```

```python
# If you want continuous index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather], axis=0)
df
```

###### Merging of DataFrame

```python
tempreture_df  = pd.DataFrame({
    'city':['mumbai','delhi','banglore','hydrabad'],
    'tempreture':[32,45,30,40]})
tempreture_df
```

```python
humidity_df = pd.DataFrame({
    'city':['delhi','mumbai','banglore'],
    'humidity':[68,65,75]})
humidity_df
```

```python
# Merge two dataframes with out explicity mention index
df = pd.merge(tempreture_df, humidity_df, on = 'city')
df
```

```python
# OUTER - JOIN
df = pd.merge(tempreture_df, humidity_df, on='city',how='outer')
df
```
